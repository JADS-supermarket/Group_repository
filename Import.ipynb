```python
import pandas as pd
import pathlib
import os

import_path = 'data/history-per-year.parquet/'
#lees alle jaren in
years = os.listdir(import_path)
for year in years:
    #lees alle maanden in per jaar (noodzakelijk om 2017 niet tot 12 maanden gaat)
    months = os.listdir(import_path+str(year)+'/')
    
    for month in months:
        data_dir = pathlib.Path(import_path+str(year)+'/'+str(month)+'/')

        #inlezen en samenvoegen tot één bestand
        df_hist = pd.concat(
            pd.read_parquet(parquet_file)
            for parquet_file in data_dir.glob('*.parquet')
        )
        #toevoegen van maand en jaar aan dataframe
        df_hist['month'] = month
    df_hist['year'] = year

df_hist.head()
```
